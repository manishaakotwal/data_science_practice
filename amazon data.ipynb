```python
# Improved function to convert salary range strings to numeric values
def parse_salary(salary_str):
    try:
        # Remove unnecessary characters
        salary_str = salary_str.replace('$', '').replace(',', '').replace('K', '000').replace('Per Hour', '').replace('(Glassdoor est.)', '').replace('(Employer est.)', '').replace('\xa0', '')
        
        # If salary is in a range (e.g., '57K - 84K'), take the average of the range
        if '-' in salary_str:
            min_salary, max_salary = salary_str.split('-')
            return (float(min_salary.strip()) + float(max_salary.strip())) / 2
        else:
            # If salary is a single value (e.g., '90K'), return it as is
            return float(salary_str.strip())
    except:
        # If parsing fails, return NaN
        return np.nan

# Apply the function to the 'Salary' column
df['Salary_numeric'] = df['Salary'].apply(parse_salary)

# Check the first few values to see if parsing worked correctly
print(df['Salary_numeric'].head())

# Calculate the average salary
average_salary = df['Salary_numeric'].mean()

# Fill null values in the 'Salary_numeric' column with the average salary
df['Salary_numeric'].fillna(average_salary, inplace=True)

# Optionally, drop the original 'Salary' column and rename 'Salary_numeric'
df.drop('Salary', axis=1, inplace=True)
df.rename(columns={'Salary_numeric': 'Salary'}, inplace=True)
```

```python
from datetime import timedelta
import pandas as pd

def custom_date_parser(date_str):
    # Check if 'd+' or 'd' is in the string and handle it
    if 'd+' in date_str or 'd' in date_str:
        days = int(date_str.replace('d+', '').replace('d', '').strip())
        return pd.Timestamp.today() + timedelta(days=days)
    try:
        # Try converting to datetime
        return pd.to_datetime(date_str)
    except ValueError:
        # Return NaT (Not a Time) if conversion fails
        return pd.NaT

# Apply the function to the 'Date' column
df['Date'] = df['Date'].apply(custom_date_parser)
```

## feature engineering

```python
# Extract the year and month from the 'Date' column and create new columns for each.
df['Date'] = pd.to_datetime(df['Date'], errors ='coerce')
# extract the year and month
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

df[['Date', 'Year', 'Month']].head()
```

```python
# Calculate the average company rating for each company.
vg_rating_per_company  = df.groupby('Company Name')['Company Rating'].agg(['mean','count'])
avg_rating_per_company
```

```python
# Sort by 'Salary' in ascending order
df_sorted_by_salary = df.sort_values(by='Salary')

# Sort by 'Date' in descending order
df_sorted_by_date = df.sort_values(by='Date', ascending=False)
```

```python
# Group by 'Job Title' and calculate the mean salary and company rating
job_title_summary = df.groupby('Job Title').agg({
    'Salary': 'mean',  # Average salary
    'Company Rating': 'mean'  # Average company rating
})

print(job_title_summary)
```

```python
# Group by 'Company Name' and count the number of job listings
company_job_counts = df.groupby('Company Name').size()

print(company_job_counts)
```

```python
# Apply multiple aggregation functions to 'Salary'
salary_stats = df['Salary'].agg(['mean', 'median', 'max'])

print(salary_stats)
```

```python
# Create a pivot table with 'Job Title' as rows, 'Company Name' as columns, and average 'Salary'
pivot_table_salary = df.pivot_table(
    values='Salary',
    index='Job Title',
    columns='Company Name',
    aggfunc='mean'
)

pivot_table_salary
```
