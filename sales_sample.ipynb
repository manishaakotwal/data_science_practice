```python
featured_outliers = ['QUANTITYORDERED','SALES','MSRP']
```

```python
# outliers removed by IQR method:
def remove_outliers(out):
    q1 = np.percentile(out, 25)
    q3 = np.percentile(out,75)

    iqr = q3 - q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr
    lower_bound, upper_bound

    out = np.where(out > upper_bound, upper_bound , np.where(out < lower_bound, lower_bound, out))
    return out[(out >= lower_bound) & (out <= upper_bound)]
```

```python
for column in featured_outliers:
    df[column] = remove_outliers(df[column])
```

```python
# outliers cleaned

int_vars = df.select_dtypes(include = ['int','float'])

# Determine the number of subplots needed
num_vars = len(int_vars.columns)
nrows = (num_vars + 2) // 3

fig, axs = plt.subplots(nrows=nrows, ncols=3, figsize=(15,10))
axs = axs.flatten()

for i, var in enumerate (int_vars):
    sns.boxplot(x=var,data=df,ax=axs[i])
    axs[i].set_title(var)

plt.tight_layout()
plt.show()
```

### 2. Sales Analysis

```python
# Total Sales per Year: Calculate the total sales per year using the YEAR_ID and SALES columns./n
# Top 5 Products by Sales: Identify the top 5 products that generated the highest sales. Use the PRODUCTCODE and SALES columns./n
# Sales by Territory: Analyze the sales distribution across different territories using the TERRITORY and SALES columns./n
# Quarterly Sales Analysis: Aggregate sales data by quarter using the QTR_ID and SALES columns to understand seasonal trends./n

total_sales_per_year = df.groupby(['YEAR_ID'])['SALES'].sum()
total_sales_per_year
```
